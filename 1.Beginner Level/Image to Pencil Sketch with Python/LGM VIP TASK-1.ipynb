### Import Packages

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

### Loading Image

```python
image = cv2.imread("jerry.jpg")
cv2.imshow("Jerry", image)
cv2.waitKey(0)
```

### Converting the Original Image to GrayScale Image:

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image,cmap="gray")
plt.title("Black And White Jerry")
```

###  Invert the GrayScale Image

```python
inverted_image = 255 - gray_image
plt.imshow(inverted_image,cmap="gray")
plt.title("Black And White Inverted")
```

### Blur the Inverted Image

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
inverted_blurred = 255 - blurred
plt.imshow(inverted_blurred,cmap="gray")
plt.title("Inverted Blurred GrayScale Image")
```

```python
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
plt.imshow(pencil_sketch,cmap="gray")
plt.title("Inverted Blurred GrayScale Image")
```

```python
cv2.imshow("Jerry Original", image)
cv2.imshow("Jerry Pencil Sketch", pencil_sketch)
cv2.waitKey(0)
```
